```python
# season 함수 정의
def get_season(date):
    if pd.Timestamp('2018-08-10') <= date <= pd.Timestamp('2019-05-12'):
        return '1819'
    elif pd.Timestamp('2019-08-09') <= date <= pd.Timestamp('2020-07-26'):
        return '1920'
    elif pd.Timestamp('2020-09-12') <= date <= pd.Timestamp('2021-05-23'):
        return '2021'
    elif pd.Timestamp('2021-08-14') <= date <= pd.Timestamp('2022-05-22'):
        return '2122'
    elif pd.Timestamp('2022-08-06') <= date <= pd.Timestamp('2023-05-28'):
        return '2223'
    elif pd.Timestamp('2023-08-11') <= date <= pd.Timestamp('2024-05-19'):
        return '2324'
    else:
        return None
    
# date를 기반으로 season 칼럼 생성
df['season'] = df['date'].apply(get_season)
```

```python
df2 = df.groupby(['domestic_competition_id', 'name', 'season','own_goals', 'opponent_goals', 'own_position', 'opponent_position', 'attendance'])['is_win'].sum().reset_index()
```
