```python
import pandas as pd

def calculate_head_to_head_results(df):
    # 결과를 저장할 빈 리스트 생성
    head_to_head_results = []

    # 각 팀마다 반복하여 상대 팀과의 전적을 계산
    for team in df['Team'].unique():
        team_data = df[df['Team'] == team]
        opponent_list = team_data['Opponent'].unique()
        
        for opponent in opponent_list:
            # 상대 팀과의 경기 결과 데이터 가져오기
            opponent_data = team_data[team_data['Opponent'] == opponent]
            result_counts = opponent_data['Result'].value_counts()
            
            # 결과를 빈 리스트에 추가
            w_count = result_counts.get('W', 0)
            d_count = result_counts.get('D', 0)
            l_count = result_counts.get('L', 0)
            
            # 새로운 행을 딕셔너리로 생성하여 리스트에 추가
            new_row = {'Team': team, 'Opponent': opponent, 'W': w_count, 'D': d_count, 'L': l_count}
            head_to_head_results.append(new_row)

    # 빈 리스트를 데이터프레임으로 변환
    head_to_head_df = pd.DataFrame(head_to_head_results)

    return head_to_head_df

# 상대 전적을 계산할 데이터프레임 준비
# 예를 들어, df는 팀과 상대팀, 그리고 결과('W', 'D', 'L')를 포함하는 데이터프레임입니다.
# 이 데이터프레임의 형식에 맞게 실제 데이터를 사용하여 함수를 호출하면 됩니다.
head_to_head_results = calculate_head_to_head_results(df)
```

```python
def calculate_head_to_head_results(df):
    # 결과를 저장할 빈 리스트 생성
    head_to_head_results_home = []
    head_to_head_results_away = []

    # 각 팀마다 반복하여 상대 팀과의 전적을 계산
    for team in df['Team'].unique():
        team_data = df[df['Team'] == team]
        opponent_list = team_data['Opponent'].unique()
        
        for opponent in opponent_list:
            # 홈 경기 결과 데이터 가져오기
            home_data = team_data[(team_data['Opponent'] == opponent) & (team_data['Venue'] == 'Home')]
            home_result_counts = home_data['Result'].value_counts()
            w_count_home = home_result_counts.get('W', 0)
            d_count_home = home_result_counts.get('D', 0)
            l_count_home = home_result_counts.get('L', 0)
            
            # 원정 경기 결과 데이터 가져오기
            away_data = team_data[(team_data['Opponent'] == opponent) & (team_data['Venue'] == 'Away')]
            away_result_counts = away_data['Result'].value_counts()
            w_count_away = away_result_counts.get('W', 0)
            d_count_away = away_result_counts.get('D', 0)
            l_count_away = away_result_counts.get('L', 0)
            
            # 홈 경기 결과를 빈 리스트에 추가
            new_row_home = {'Team': team, 'Opponent': opponent, 'Venue': 'Home', 'W': w_count_home, 'D': d_count_home, 'L': l_count_home}
            head_to_head_results_home.append(new_row_home)
            
            # 원정 경기 결과를 빈 리스트에 추가
            new_row_away = {'Team': team, 'Opponent': opponent, 'Venue': 'Away', 'W': w_count_away, 'D': d_count_away, 'L': l_count_away}
            head_to_head_results_away.append(new_row_away)

    # 빈 리스트를 데이터프레임으로 변환
    head_to_head_df_home = pd.DataFrame(head_to_head_results_home)
    head_to_head_df_away = pd.DataFrame(head_to_head_results_away)

    return head_to_head_df_home, head_to_head_df_away

# 상대 전적을 계산할 데이터프레임 준비
# 예를 들어, df는 팀과 상대팀, Venue 그리고 결과('W', 'D', 'L')를 포함하는 데이터프레임입니다.
# 이 데이터프레임의 형식에 맞게 실제 데이터를 사용하여 함수를 호출하면 됩니다.
head_to_head_results_home, head_to_head_results_away = calculate_head_to_head_results(df)
```
